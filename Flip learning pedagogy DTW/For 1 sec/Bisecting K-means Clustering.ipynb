```python
import pandas as pd
import numpy as np
from tslearn.metrics import dtw
import math
ground_truth_paths=[]
test_sets_path=[]
for i in range(1,21):
    #path of ground truth for ith student file
    path1="C:\\Users\\KIIT\\Downloads\\Flip learning pedagogy DTW\\Student_20\\Student_20\\S"+str(i)+".csv"
#     print(path1)
    ground_truth_paths.append(path1)

for i in range(1,21):
    a=[]
    for j in range(10):
        #path of jth video EEG file for ith student
        path2="C:\\Users\\KIIT\\Downloads\\Flip learning pedagogy DTW\\DataSet\\DataSet_one_second\\S"+str(i)+"_1\\s"+format(i,'02d')+"_video"+str(j)+".csv"
#         print(path2)
        a.append(path2)
    test_sets_path.append(a)
DTW_dist_score=[]
for i in range(20):
    df1 = pd.read_csv(ground_truth_paths[i], header=None)
    ts1=df1.to_numpy()
#     ts1 = ts1.astype(float)
    mean1 = ts1.mean(axis=0, keepdims=True)
    std1 = ts1.std(axis=0, keepdims=True)
    ts1_norm = (ts1 - mean1) / std1
    
#     print(df1.to_string())
    a=[]
    for j in range(10):
        df2 = pd.read_csv(test_sets_path[i][j], header=None)
#         print(df2.to_string())
        # Convert each DataFrame to a time series
        ts2 = df2.to_numpy()
        # Normalize the time series        
        mean2 = ts2.mean(axis=0, keepdims=True)
        std2 = ts2.std(axis=0, keepdims=True)
        ts2_norm = (ts2 - mean2) / std2

        # Calculate the DTW similarity score
        dist = dtw(ts1_norm, ts2_norm)
        a.append(dist)
        print(dist,end=" ")
    DTW_dist_score.append(a)
    print()
similarity_score=[]
print("*************")
min_DTW=min([min(r) for r in DTW_dist_score])
max_DTW=max([max(r) for r in DTW_dist_score])
print(min_DTW)
print(max_DTW)
dissimilarity_score=[]
for i in range(len(DTW_dist_score)):
    a=[]
    b=[]
    for j in range(len(DTW_dist_score[0])):
        dissimilarity=(DTW_dist_score[i][j] - min_DTW) / (max_DTW - min_DTW)
        print(dissimilarity,end=" ")
        b.append(dissimilarity)
        if(dissimilarity<=0.4):
            a.append(1)
        else:
            a.append(0)
    similarity_score.append(a)
    dissimilarity_score.append(b)
    print()
print("DTW similarity score:")
print(similarity_score)
print(dissimilarity_score)

```
