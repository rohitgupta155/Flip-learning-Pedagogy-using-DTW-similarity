```python
import pandas as pd
import numpy as np
from tslearn.metrics import dtw
import math
ground_truth_paths=[]
test_sets_path=[]
for i in range(1,21):
    #path of ground truth for ith student file
    path1="C:\\Users\\KIIT\\Downloads\\Minor Project\\Student_20\\Student_20\\S"+str(i)+".csv"
#     print(path1)
    ground_truth_paths.append(path1)

for i in range(1,21):
    a=[]
    for j in range(10):
        #path of jth video EEG file for ith student
        path2="C:\\Users\\KIIT\\Downloads\\Minor Project\\DataSet\\DataSet_one_second\\S"+str(i)+"_1\\s"+format(i,'02d')+"_video"+str(j)+".csv"
#         print(path2)
        a.append(path2)
    test_sets_path.append(a)
DTW_dist_score=[]
for i in range(20):
    df1 = pd.read_csv(ground_truth_paths[i], header=None) 
    ts1 = df1.to_numpy()
    mean1 = ts1.mean(axis=1, keepdims=True)
    std1 = ts1.std(axis=1, keepdims=True)
    ts1_norm = (ts1 - mean1) / std1
#     print(df1.to_string())
    a=[]
    if i==11:
        continue
    for j in range(10):
        df2 = pd.read_csv(test_sets_path[i][j], header=None)
#         print(df2.to_string())
        # Convert each DataFrame to a time series
        ts2 = df2.to_numpy()

        # Normalize the time series
        mean2 = ts2.mean(axis=1, keepdims=True)
        std2 = ts2.std(axis=1, keepdims=True)
        ts2_norm = (ts2 - mean2) / std2

        # Calculate the DTW similarity score
        dist = dtw(ts1_norm, ts2_norm)
        a.append(dist)
        print(dist,end=" ")
    DTW_dist_score.append(a)
    print()
similarity_score=[]
print("*************")
min_DTW=min([min(r) for r in DTW_dist_score])
max_DTW=max([max(r) for r in DTW_dist_score])
print(min_DTW)
print(max_DTW)
for i in range(len(DTW_dist_score)):
    a=[]
    for j in range(len(DTW_dist_score[0])):
        similarity=(DTW_dist_score[i][j] - min_DTW) / (max_DTW - min_DTW)
        print(similarity,end=" ")
        a.append(round(1-similarity))
    similarity_score.append(a)
    print()
print("DTW similarity score:")
print(similarity_score)

```

```python
import pandas as pd
import numpy as np
from tslearn.metrics import dtw
import math
from dtwParallel import dtw_functions
from scipy.spatial import distance as d
ground_truth_paths=[]
test_sets_path=[]
for i in range(1,21):
    #path of ground truth for ith student file
    path1="C:\\Users\\KIIT\\Downloads\\Minor Project\\Student_20\\Student_20\\S"+str(i)+".csv"
#     print(path1)
    ground_truth_paths.append(path1)

for i in range(1,21):
    a=[]
    for j in range(10):
        #path of jth video EEG file for ith student
        path2="C:\\Users\\KIIT\\Downloads\\Minor Project\\DataSet\\DataSet_one_second\\S"+str(i)+"_1\\s"+format(i,'02d')+"_video"+str(j)+".csv"
#         print(path2)
        a.append(path2)
    test_sets_path.append(a)
DTW_dist_score=[]
for i in range(20):
    df1 = pd.read_csv(ground_truth_paths[i], header=None) 
    ts1 = df1.to_numpy()
    a=[]
    if i==11:
        continue
    for j in range(10):
        df2 = pd.read_csv(test_sets_path[i][j], header=None)
        ts2 = df2.to_numpy()
        dist=dtw_functions.dtw(ts1, ts2, type_dtw="d", local_dissimilarity=d.euclidean, MTS=True)
        a.append(dist)
        print(dist,end=" ")
    DTW_dist_score.append(a)
    print()
print(DTW_dist_score)
similarity_score=[]
print("*************")
min_DTW=min(min(DTW_dist_score))
max_DTW=max(max(DTW_dist_score))
print(min_DTW)
print(max_DTW)
for i in range(len(DTW_dist_score)):
    a=[]
    for j in range(len(DTW_dist_score[0])):
        similarity=(DTW_dist_score[i][j] - min_DTW) / (max_DTW - min_DTW)
        print(similarity,end=" ")
        a.append(round(1-similarity))
    similarity_score.append(a)
    print()
print("DTW similarity score:")
print(similarity_score)
```

```python
def dtw_distance(s1, s2):
    # Initialize matrix
    DTW = np.zeros((len(s1) + 1, len(s2) + 1))
    for i in range(len(s1) + 1):
        for j in range(len(s2) + 1):
            DTW[i][j] = float('inf')
    DTW[0][0] = 0

    # Fill matrix
    for i in range(1, len(s1) + 1):
        for j in range(1, len(s2) + 1):
            cost = abs(s1[i-1] - s2[j-1])
            DTW[i][j] = cost + min(DTW[i-1][j], DTW[i][j-1], DTW[i-1][j-1])

    # Return DTW score
    return DTW[-1][-1]
DTW_dist_score=[]
for i in range(20):
    df1 = pd.read_csv(ground_truth_paths[i], header=None) 
    ts1 = df1.to_numpy()
    a=[]
    if i==11:
        continue
    for j in range(10):
        df2 = pd.read_csv(test_sets_path[i][j], header=None)
        ts2 = df2.to_numpy()
        dist=dtw(ts1, ts2)
        a.append(dist)
        print(dist,end=" ")
    DTW_dist_score.append(a)
    print()
print(DTW_dist_score)
similarity_score=[]
print("*************")
min_DTW=min(min(DTW_dist_score))
max_DTW=max(max(DTW_dist_score))
print(min_DTW)
print(max_DTW)
for i in range(len(DTW_dist_score)):
    a=[]
    for j in range(len(DTW_dist_score[0])):
        similarity=(DTW_dist_score[i][j] - min_DTW) / (max_DTW - min_DTW)
        print(similarity,end=" ")
        a.append(round(1-similarity))
    similarity_score.append(a)
    print()
print("DTW similarity score:")
print(similarity_score)
```
